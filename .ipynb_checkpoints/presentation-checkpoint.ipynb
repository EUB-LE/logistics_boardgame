# Evaluation: Boardgame with Simple Agents


```python
from boardgame.classes import *
from boardgame.agent import Agent
import pandas as pd
import numpy as np
from collections import Counter
```

```python
def parse_result_list(groups: pd.core.groupby.DataFrameGroupBy) -> pd.DataFrame:
    """
    calculate aggregated values from a list of game results, grouped by the desired attribute
    """
    return groups.agg(
    observations=("turn", "count"),
    avg_turns=("turn", "mean"),
    min_turns=("turn", "min"),
    max_turns=("turn", "max"),
    win_percentage=("result", lambda x : Counter(x)["WON"] / len(x) * 100),
    lost_percentage=("result", lambda x : Counter(x)["LOST"] / len(x) * 100), 
    cascade_percentage=("reason", lambda x: Counter(x)["CASCADE"] / len(x) * 100), 
    no_cards_left_percentage=("reason", lambda x: Counter(x)["PLAYER-CARDS"] /len(x) * 100)
)
```

## Results 

### Maximal Cascade Level 
The win percentage increases as the game allows more cascades to take place before it is lost. The losing conditions at extremely high max values (100,200,500,1000,2000,3000) show us that if all other game parameters stay unchanged, a cascade will always take place before the player cards run out. The following analysis will therefore assume that a game was lost because of too many cascade unless stated otherwise.

```python
results = []
wins_and_losses = []
for cascade_max_level in [i for i in range(1,40,5)]+[100,200,500,1000,2000,3000]: 
    for random_seed in range(0,100):
        g = Game(random_seed=random_seed, cascade_max_level=cascade_max_level)
        g.set_agent(Agent(g))
        result = g.play_game()
        result['cascade_max_level'] = cascade_max_level
        wins_and_losses.append(result)
```

### Cascade Damage Threshold

```python
wins_and_losses2 = []
for cascade_damage_threshold in [i for i in range(1,10,1)]: 
    for random_seed in range(0,100):
        g = Game(random_seed=random_seed, cascade_damage_threshold=cascade_damage_threshold)
        g.set_agent(Agent(g))
        result = g.play_game()
        result['cascade_damage_threshold'] = cascade_damage_threshold
        wins_and_losses2.append(result)
```

```python
df2 = pd.DataFrame(wins_and_losses2)
damage_threshold_result = parse_result_list(df2.groupby("cascade_damage_threshold"))
damage_threshold_result
```

```python
damage_threshold_result.plot.bar(y=["win_percentage", "avg_turns"])
```

### Target Amount and Cascade Damage Threshold

```python
wins_and_losses3 = []
for target_amount in [i for i in range(2,7)]: 
    for cascade_damage_threshold in range(2,6):
        for random_seed in range(0,100):
            g = Game(random_seed=random_seed, disable_logging = True, target_amount=target_amount, cascade_damage_threshold=cascade_damage_threshold)
            g.set_agent(Agent(g))
            result = g.play_game()
            result['target_amount'] = target_amount
            result['cascade_damage_threshold']=cascade_damage_threshold
            wins_and_losses3.append(result)
```

```python
df3 = pd.DataFrame(wins_and_losses3)
amount_result = parse_result_list(df3.groupby(["target_amount", "cascade_damage_threshold"]))
amount_result
```

```python
# get unique cascade_damage_thresholds
temp_dict = {}
for x in list(dict.fromkeys([x[1] for x in amount_result.index])):
    temp_dict[f"damage threshold {x}"]=(amount_result.xs(x, level='cascade_damage_threshold')["win_percentage"])


ax = pd.DataFrame(temp_dict).plot.bar()
ax.set_ylabel("win percentage")
```
